# Data Cleaning

```python
date = df.Date.str.split('-', expand =True)
df['year'] = date[2]
df['month'] = date[1]
df['day'] = date[0]
df.year = pd.to_numeric(df.year)
df.month = pd.to_numeric(df.month)
df.day = pd.to_numeric(df.day)
```

```python
df.head()
```

# Q.1 What is the overall trend of the stock market?




```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df.index, y=df['Listing_Close'])
plt.title('Overall Trend of Nifty 50')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()
```

# Q.2 What is the distribution of daily returns for the stock market index?

```python
df['Daily_Return'] = df['Listing_Close'].pct_change() * 100
```

```python
df = df.dropna()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Daily_Return'], bins=50, kde=True)
plt.title('Distribution of Daily Returns for Nifty 50')
plt.xlabel('Daily Return (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Daily_Return'])
plt.title('Box Plot of Daily Returns for Nifty 50')
plt.ylabel('Daily Return (%)')
plt.show()
```

# Q.3 What are the highest and lowest daily returns?

```python
df['Daily_Return'] = df['Listing_Close'].pct_change() * 100
df = df.dropna()
```

```python
highest_return = df['Daily_Return'].max()
lowest_return = df['Daily_Return'].min()

print(f"Highest Daily Return: {highest_return:.2f}%")
print(f"Lowest Daily Return: {lowest_return:.2f}%")
```

```python
plt.figure(figsize=(8, 6))
plt.bar(['Highest', 'Lowest'], [highest_return, lowest_return], color=['green', 'red'])
plt.title('Highest and Lowest Daily Returns')
plt.ylabel('Daily Return (%)')
plt.show()
```

#Q.4 Average daily return by month

```python
df["Month"] = df.index.month
```

```python
monthly_returns = df.groupby('Month')['Daily_Return'].mean()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_returns.index, y=monthly_returns.values)
plt.title('Average Daily Returns by Month')
plt.xlabel('Month')
plt.ylabel('Average Daily Return (%)')
plt.show()
```

# Q.5 What is the average daily return for each year?

```python
yearly_returns = df.groupby(df.index.year)['Daily_Return'].mean()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=yearly_returns.index, y=yearly_returns.values)
plt.title('Average Daily Returns by Year')
plt.xlabel('Year')
plt.ylabel('Average Daily Return (%)')
plt.show()
```
